## Question 2
#### Explain how patches relate to the idea of committing a change to the history of your code?

````python
While a patch is the description of a change in Git (a diff), a commit is a snapshot of the repository 
that becomes part of the code history. Staging edited code using `git add` allows the user to see the difference between past versions 
of the code and the patched versions using `git diff --staged`. Using `git commit` then takes the content that has been staged and records it 
as a snapshot in the repository. Since Git stores snapshots instead of diffs, a patch is a convinient way to represent and move changes between snapshots.
This makes patches extremely useful for review of your code, especially for changes across branches.
````

## Question 3
#### What is the difference between a branch and a fork?

````python
A branch is a pointer inside a single Git repository; creating a branch simply creates a new ref that points
to a commit, making it easier to develop a feature or fix a bug while keeping the main unchanged. Branches are
usually merged back into the original repository when finished working. 
    A fork is a server-sided copy of an entire repository made by a different user. It gives the user an independent
copy that they control with their own permissions, and allows them to contribute to projects when they otherwise lack
push access to the upstream repo.
    Common workflow is to create a branch for a single unit of work, push the branch to your fork and then open a pull
request from your-fork/your-branch into upstream/main. Branches allow users to keep multiple workstreams in the same repo,
while forks allow for an isolated workspace to host those branches and sync with the upstream repo to submit contributions.
````

## Question 4
#### How are diff and patch related? What’s the difference?

````python
Diff is the human-readable, textual representation of changes between two file states. A patch is a formatted diff that can be applied
to another working tree, imported, or committed. While they may be used interchangeably in conversation, it is important to note 
the difference: `git diff` produces a diff you can read to see added/removed lines, while `git format-patch (or `git diff > file.patch
or `.patch`/`.diff`) produces a patch file that can be used to reproduce the change in another repo or point in code history. Diff and patch 
are related in that they are ways to handle commits. In the Git workflow, a diff (like in a pull request) can be used to review changes 
and then the repository records the result as commits. Exporting commits as patches explicitly packages those diffs so that they can be 
applied to another working tree or shared later.
````

## Question 5
#### See the diagram below.
```
alice            *-*-*-*-d-e     
bob              *-*-*-*-a-b-c
origin/main      *-*-*-*-a-b-c
```
#### a) What part of the Git trininty does each line represent?
#### b) What would the result be if Alice performed a rebase? Edit the diagram. What would she type in the terminal to rebase, then push?
#### c) What would the result be if Alice performed a merge? Edit the diagram. What would she type in the terminal to merge, then push?
#### d) Which would be the better option (rebase vs. merge)? Why?

````python
(a) What part of the Git trinity does each line represent?
Each line represents a branch reference that names a place in the repo`s commit graph. 
origin/main is the remote-tracking branch, Bob is another local branch that points at the same
commits as the origin/main, and alice is another local branch with different commits than the main.
Each * and letter represents commits along the branches, with the letters being branch-specific commits.
    
(b) What would the result be if Alice performed a rebase? Edit the diagram. What would she type to rebase, then push?
If Alice performed da rebase onto origin/main, Git will replay her changes (d and e) on top of a-b-c. That produces new commits for her changes:
alice            *-*-*-*-a-b-c-d-e
bob              *-*-*-*-a-b-c
origin/main      *-*-*-*-a-b-c

Commands Alice would run:
git fetch origin
git checkout alice
git rebase origin/main
git push --force-with-lease origin alice

(c) What would the result be if Alice performed a merge? Edit the diagram. What would she type in the terminal to merge, then push?
If Alice merges origin/main into her branch, Git will create a merge commit that has two parents (one pointing to her e commit 
and one pointing to c). The history will contain both lines and then a merge commit M:
alice            *-*-*-*-d-e-M
bob              *-*-*-*-a-b-c
origin/main      *-*-*-*-a-b-c

Commands Alice would run:
git fetch origin
git checkout alice
git merge origin/main        # resolve conflicts if any, then git commit finishes the merge
git push origin alice

(d) Which would be the better option (rebase vs. merge)? Why?
In Alice's case, since she is working on a team, merge may be better. In order to preserve the true chronological branch structure,
merge is better as it is non-destructive and preserves Alices record in parallel to the main. Merge commit avoids the need to force-push
and keeps transparency. Alice may want to keep a clean branch before merging into upstream, but rebase rewrites history and could
cause disruption if anyone based their work on her branch.
````
